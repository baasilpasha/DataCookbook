# Chapter 3 - Data Exploration
Exploring the data involves visualizing it and understanding how values are laid out. We can identify patterns, outliters and overall shapes of the data with this.

```python
import pandas as pd
import matplotlib.pyplot as plt #important plotting library
import seaborn as sns #important visualization library
df = pd.read_csv('Electric_Vehicle_Population_Data.csv')
```

## 3.1 Histogram
One way to look at the data is to see what the distribution of the Electric Range of the cars is:

```python
plt.figure(figsize=(10, 6)) #size of figure
sns.histplot(df['Electric Range'], bins=20) #histogram plot, there are a lot of parameters to play with
plt.title('Distribution of Electric Range') #title
plt.xlabel('Electric Range') #axis labels
plt.ylabel('Frequency')
plt.show()
```

The distribution almost doesn't make sense, why would so many cars have such low range? But we have to remember that the dataset also includes information on hybrid cars. This is why exploring the data is neccesary, so we understand what we need to manipulate to get our desired results.

There is also another way to create the histogram, without using Seaborn, and just matplot lib directly:

```python
plt.hist(df['Electric Range'], bins = 20)

plt.xlabel('Electric Range')
plt.ylabel('Frequency')
plt.title('Histogram of Electric Range')
```
